```python
%pylab inline
import pandas as pd
import seaborn

f_data = open( 'plate1_data.txt' )
f_map  = open( 'plate1_map.txt'  )

data = []
for line_data in f_data :
    for line_map in f_map :
        cells_data = line_data.strip().split('\t')
        cells_map  = line_map.strip().split('\t')
        data = data + zip( cells_map, cells_data)

categoreis = [ 'Bug', 'Temp', 'Substrate', 'OD' ]
table = {}
for n,cell in enumerate(data) :
    if cell[0] == 'EMPTY' : continue
    table[n] = dict(zip( categoreis, cell[0].split(', ') + [ cell[1] ] ))
    table[n]['Temp'] = table[n]['Temp'] + 'C'
temp_table = pd.DataFrame( table ).T
temp_table.head()
```
